# Plots by Group 

# Seaborn for Faceted plots/ Plotting by groups

```python
data.Department.value_counts()
```

```python
sns.FacetGrid(col='Department',col_wrap=25,aspect=.8,size=4,data=data).map(plt.scatter, 'Age', 'Income')
```

```python
sns.FacetGrid(col='MaritalStatus',col_wrap=25,aspect=.8,size=4,data=data).map(plt.hist,  'Income')
```

```python
# Histogram of Income by Marital Status
sns.FacetGrid(col='MaritalStatus',row = 'Gender',aspect=.8,size=4,data=data).map(plt.hist,  'Income')
```

# Using Subplots for multiple plots - matplot lib

```python
# Plot 1 : Scatter Plot with Age and Income
plt.subplot(1, 2, 1)

plt.scatter(data=data, x = 'Age', y='Income')

# Plot 2 : Box Plot with Income
plt.subplot(1, 2, 2)

plt.boxplot(data['Income'])

# Look Good
plt.tight_layout()
```
